```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv(r'Q:\Users\adria\PycharmProjects\Personal\Projects\ClimateAnalysisModel\data\clean_data\clean_data.csv')
print(df.head())
print(df.describe())
```

```python
annual_avg_temp = df.groupby('start')['tavg'].mean()
print(annual_avg_temp.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(annual_avg_temp.index, annual_avg_temp.values, label='Annual Average Temperature', color='blue', marker='o')
plt.title('Annual Average Temperature Over Time')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.legend()
plt.show()
```

```python
result = seasonal_decompose(annual_avg_temp, model = 'additive', period=1)
result.plot()
plt.show()
```

```python
annual_avg_tmin = df.groupby('start')['tmin'].mean()
annual_avg_tmax = df.groupby('start')['tmax'].mean()

plt.figure(figsize=(12, 6))
plt.plot(annual_avg_tmin.index, annual_avg_tmin.values, label='Average Minimum Temperature', color='skyblue', marker='o')
plt.plot(annual_avg_tmax.index, annual_avg_tmax.values, label='Average Maximum Temperature', color='salmon', marker='o')
plt.title('Average Temperature Extremes Over Time')
plt.xlabel('Start Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```
